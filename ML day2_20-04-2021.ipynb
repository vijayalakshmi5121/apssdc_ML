```python
input_data=head_df.drop('Brain Weight(grams)',axis=1)
output_data=head_df['Brain Weight(grams)']
```

```python
### Applying the model
from sklearn.linear_model import LinearRegression
linear=LinearRegression()
```

```python
## Train the model by using fit method
linear.fit(input_data,output_data)
```

```python
pred=linear.predict(input_data)
pred
```

```python
## Evaluate the Model
linear.score( input_data,output_data)*100
```

### Linear Regression with Applestore.csv

```python
apple_df=pd.read_csv('https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Regression/AppleStore.csv')
```

```python
apple_df.head()
```

```python
apple_df.columns
```

```python
apple_df.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
lbr=LabelEncoder()
```

```python
apple_df.info()
```

```python
apple_df['currency']=lbr.fit_transform(apple_df['currency'])
apple_df['ver']=lbr.fit_transform(apple_df['ver'])
apple_df['cont_rating']=lbr.fit_transform(apple_df['cont_rating'])
apple_df['prime_genry']=lbr.fit_transform(apple_df['prime_genre'])
```

```python
apple_df.drop(['Unnamed: 0','track_name'],axis=1,inplace=True)
```

```python
apple_df.head()
```

```python
x=apple_df.drop('price',axis=1)
y=apple_df['price']
```

```python
sns.pairplot(apple_df)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
```

```python
y_pred=lr.predict(x)
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```
